```python
from sklearn.preprocessing import StandardScaler

def plot_timeseries(df, index='Time', isScaled = False):
    cols_to_plot = []
    for col in df.columns:
        if col != 'Time':
            cols_to_plot.append(col)
            
    if isScaled:
        df[cols_to_plot] = StandardScaler().fit_transform(df[cols_to_plot])
    df.plot.line(x=index, y=cols_to_plot)
    plt.show()
    
plot_timeseries(df, isScaled=True)
```

### Modelling

Once we have performed some feature analysis and engineering, we can now move the modelling stages of the Data Science process. 

Before jumping into complex methods and modelling approaches, it is important to fiurst establish your baseline mode, which would be using a very simple approach first, and then work towards more complex models, and even, more complex combinations of classifiers in order to build a robust predictive model.

The first step here will be to prepare our data, which involves converting our dataframes into matrices, and also (if being used in a supervised learning model), splitting the data into train and test (and if required), validate.

Many times, the methodology of splitting data is extremely important in the training and validation of a model. Take a timeseries model, when performing modelling, one would to preserve the order of the data points, as the sequence is an important characteristic of a timeseires model (e.g. ARIMA). 

In a classification task, sampling techniques is also highly important, as if we are dealing with a imbalanced set of classes (e.g. 90% of the data points are for 1 class out of k), then if splitting the data without awaremess of the proportion of data points within a given class, we will affect the models capability to provide a suitable decision boundary / H(x). In these cases, we start to encounter forms of overfitting and underfitting.

It is important to also consider how one can perform cross-validation on the data in order to reduce the chances of badly performing models.

**Cross Validation**
Cross-Validation is similar to train/test split, but its applied to more subsets. Meaning, the data is split into k subsets, and train on k-1 one of those subset. What we do is to hold the last subset for test. Were able to do it for each of the subsets.

### Deep Neural Network (TensorFlow - Low Level Implementation)

A **Deep Neural Network** (DNN) is an artificial neural network (ANN) with multiple layers between the input and output layers. The DNN finds the correct mathematical manipulation to turn the input into the output, whether it be a linear relationship or a non-linear relationship.

**TensorFlow**  is an end-to-end open source platform for machine learning. It has a comprehensive, flexible ecosystem of tools, libraries and community resources that lets researchers push the state-of-the-art in ML and developers easily build and deploy ML powered applications.

Tensorflow allows you to construct Neural Networks using an API-like design, to more lower-level architectures, which require deeper knowledge of how a machine learning model computes the hypothesis and cost funtions. Whilst this is a little more complicated, it's good to understand the internal workings of a DNN before working with the higher-level APIs.

For our implementation, we're going to work with the lower-level implementation first, and develop a DNN which performs 2-class classifications, and is sensitive to imbalance classes.

**Network Architecture**

Traditionally, neural networks only had three types of layers: hidden, input and output. These are all really the same type of layer if you just consider that input layers are fed from external data (not a previous layer) and output feed data to an external destination (not the next layer).

### TensorFlow - Autocoder Approach

The autoencoder tries to learn a function hW,b(x)x. In other words, it is trying to learn an approximation to the identity function, so as to output x  that is similar to x. The identity function seems a particularly trivial function to be trying to learn; but by placing constraints on the network, such as by limiting the number of hidden units, we can discover interesting structure about the data.
